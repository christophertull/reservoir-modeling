```python
import pandas as pd
import statsmodels.tsa.api as tsa
import matplotlib.pyplot as plt
from datetime import datetime
import numpy as np
%matplotlib inline
```

```python
data = pd.read_csv('data\\ts_ORO.csv', parse_dates=True, index_col=0)
testDate = datetime(2013,11,1)
```

```python
model = tsa.VAR(data[:testDate])
```

```python
result = model.fit()
result.summary()
```

```python
result.plot_acorr()
```

```python
result.plot_forecast(100)
```

```python
params = result.params
coeffs = params.STORAGEAF
coeffs
```

```python
prev_day_storage = data.STORAGEAF[testDate]
pred_storage = []

for day in data[testDate:].iterrows():
    day = day[1]
    pred =  coeffs[0] + coeffs[1]*day[0] + prev_day_storage + coeffs[3]*day[2] + coeffs[4]*day[3]
    pred_storage.append(pred)
    prev_day_storage = pred
    
    
print len(pred_storage)
```

```python
testModel = tsa.VAR(data[testDate:])
# print data[testDate:]
predictions = testModel.predict(params)[:,1]

tmp = pd.DataFrame()
tmp['Actual'] = data[testDate:].STORAGEAF[1:]
# tmp['Predicted'] = predictions
tmp['Predicted'] = pred_storage[:-1]

ax = tmp.plot(figsize=(10,8))
ax.set_title("Resevoir Storage")
ax.set_xlabel('Month')
ax.set_ylabel('Volume (Acre-Feet)')
```

```python
tmp['Error'] = tmp.Actual-tmp.Predicted
ax = tmp.Error.plot(figsize=(10,8))
ax.set_title("Prediction Error")
ax.set_xlabel('Month')
ax.set_ylabel('Volume (Acre-Feet)')
```

```python
tmp['% Error'] = (tmp.Actual-tmp.Predicted)/tmp.Actual
ax = tmp['% Error'].plot(figsize=(10,8))
ax.set_title("Prediction Error")
ax.set_xlabel('Month')
ax.set_ylabel('Volume (Acre-Feet)')
```
